```python
import matplotlib.pyplot as plt
plt.hist(iris['species'])
```

### 문제 [2-2] (10점)

### 문제 [2-3] (10점)

iris의 species 열 값을 setosa는 1, 그 외의 값들은 0으로 변환하시오. 
(apply, lambda 사용)

```python
iris['species'] = iris.apply(lambda x: 1 if x['species'] == 'setosa' else 0, axis = 1)
iris
```

# 문제 3

다음 데이터에 사용하여 각 문항에 따라 답하시오.

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

### 문제 [3-1] (10점)

titanic에서 pclass, age, sibsp, fare, class 열만 뽑아 df에 저장하시오.

```python
df = titanic.loc[:, ['pclass', 'age','sibsp', 'fare', 'class' ]].iloc[:,:];df
```

### 문제 [3-2] (5점)

df의 열 별 결측치 개수를 구하시오.

```python
df.isnull().sum()
```

### 문제 [3-3] (10점)

df에 결측치가 있다면, 해당 열의 평균값으로 채우시오.

```python
df.loc[df['age'] != df['age'], 'age'] = df['age'].mean()
df
```

### 문제 [3-4] (5점)

survived.csv를 불러와 survived에 저장하시오.

```python
survived = pd.read_csv('C:/Users/USER-PC/anaconda3/WorkSpace/Excel 파일/survived.csv')
survived
```

### 문제 [3-5] (10점)

기존에 생성한 df와 survived 데이터프레임을 병합해 df2에 저장하시오.

### 문제 [3-6] (20점)

df2 데이터에서 각 class 별 생존율을 구하시오.

```python
print(df2.groupby(['class','survived']).count())

df2[['class', 'survived']].groupby('class').mean()
```
