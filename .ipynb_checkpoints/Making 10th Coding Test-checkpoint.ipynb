결과값으로 점수가 높은 5명의 점수를 출력하시오

리스트 순서대로 1명의 학생의 점수입니다

아래 코드를 실행시키고 문제에 답하시오.

hint) for문 사용

```python
english = [84, 76, 86, 90, 73, 85, 81, 71, 89, 100]
math = [72, 76, 86, 71, 69, 96, 74, 83, 87, 68]

# 위에는 주어짐

total = []

'''
for i, j in zip(english, math):
    total.append(i + j)
'''

 
for i in range(10):
    total.append(english[i] + math[i])

total.sort(reverse = True)

print(total[0:5])
```

[문제 3-4]
county가 'SCOTT'인 경우와 country가 'PIATT'인 경우의 poptotal의 평균을 각각 구하시오.
hint) query()와 mean()

```python
SCOTT = midwest.query('county == "SCOTT"')['poptotal'].mean()
PIATT = midwest.query('county == "PIATT"')['poptotal'].mean()
print(SCOTT)
print(PIATT)
```
