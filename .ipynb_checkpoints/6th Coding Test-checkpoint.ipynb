# 문제 3.
score는 학생들의 점수이다. 반복문과 조건문을 사용하여 점수가 80점 이상이면 우수, 60점 이상 80점 미만은 보통 , 그렇지 않으면 미흡임을 알려주도록 하시오.

```python
score = [90,25,67,45,80]  

for i in range(5):
    if score[i] >= 80:
        print('우수')
    elif score[i] >= 60:
        print('보통')
    else:
        print('미흡')
```
